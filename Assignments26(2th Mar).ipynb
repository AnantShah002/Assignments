### Q2) What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Ans) **Scatter plot :-** Scatter plot connest x and y point thein plot like it shows the relationship between two variables

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
x,y
```

```python
plt.scatter(x,y)
plt.title("This is our X & Y Data")
plt.xlabel("X Data")
plt.ylabel("Y Data")
```

### Q3) Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

#### Ans) **subplot():-** function adds subplot to a current figure at the specified grid position.  It is similar to the subplots() function however unlike subplots() it adds one subplot at a time. So to create multiple plots you will need several lines of code with the subplot() function.

```python
fig, ex = plt.subplots(4, 1)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
ex[0].plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
ex[1].plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
ex[2].plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
ex[3].plot(x,y)
```

### Q4) What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

#### Ans) **Bar plot** a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```
